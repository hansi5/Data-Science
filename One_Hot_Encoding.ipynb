# Import libraries

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

# Data file

```python
df = pd.read_excel('car_price.xlsx')
df
```

## Dummy columns

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
merged = pd.concat([df, dummies], axis='columns')
final_df = merged.drop(['Car Model', 'Mercedez Benz C class'], axis='columns')
final_df
```

```python
x = final_df.drop(['Sell Price($)'], axis='columns')
y = final_df['Sell Price($)']
y
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(x, y)
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.score(x, y)
```

## One Hot Encoder

```python
le = LabelEncoder()
```

```python
df['Car Model'] = le.fit_transform(df['Car Model'])
df
```

```python
x = df[['Car Model', 'Mileage', 'Age(yrs)']]
y = df['Sell Price($)']
x
```

```python
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([("Car Model", OneHotEncoder(), [0])], remainder = 'passthrough')
X = ct.fit_transform(x)[:, 1:]
X
```

```python
model.fit(X, y)
```

```python
model.predict([[0, 1, 45000, 4]])
```

```python
model.score(X, y)
```
